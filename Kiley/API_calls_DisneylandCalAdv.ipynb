```python
# Pretty Print the output of the JSON
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
from pprint import pprint
import pandas as pd

parks = response["lands"]

data = []

for park in parks:
    print("Next Park")
    pprint(park["name"])
    rides = park["rides"]
    pprint(rides)
    #use a for loop to iterate through 'rides' and print the names
    for ride in rides:
        ride_info = {
            "Park": park["name"],
            "Ride Name": ride["name"],
            "Is Open": ride["is_open"],
            "Wait Time": ride["wait_time"],
            "Last Updated": ride["last_updated"]
        }
        data.append(ride_info)

        #here is where you would insert it as a dataframe or add to a csv that you have opened
        df = pd.DataFrame(data)
        #dataframe row 
        csv_filename = "DisneylandCalAdv_rides_data.csv"
        df.to_csv(csv_filename, index=False)
        # Create a DataFrame
        df = pd.DataFrame(data)
```
